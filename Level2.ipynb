Task 3:  Geographic Analysis

(A)  Plot the locations of restaurants on a map using longitude and latitude coordinates.

```python
#Ploting the location of restaurants on a map using longitude and latitude coordinates:
import folium 
from folium.plugins import MarkerCluster
m=folium.Map(location=[df.loc[0,'Latitude'],df.loc[0,'Longitude']],zoom_start=6)

marker_cluster=MarkerCluster().add_to(m)

for series in df.itertuples():
    folium.Marker([series.Latitude,series.Longitude],popup=series.City,tooltip=series.City,icon=folium.Icon(icon='cutlery',color='red')).add_to(marker_cluster)

m
```

This map show different location of restaurants in each city of country by latitude and longitude.

(B)  Analyze the ratings and popularity of different restaurant chains.

 Top 2 restaurant chains rating and popularity :

1. Cafe Coffee Day

```python
#Rating and popularity of different restaurant chains :

#Cafe Coffee Day Restaurant 

#In City:

restaurant_Name=df[df['Restaurant Name']=='Cafe Coffee Day']

# Restaurant chains rating count
rating_Restaurant_Chains1=restaurant_Name.groupby(df[df['City']=='New Delhi']['City'])[['Restaurant Name','Aggregate rating']].value_counts()

# Restaurant chain rating color count
color_Restaurant_Chains1=restaurant_Name.groupby(df[df['City']=='New Delhi']['City'])[['Restaurant Name','Rating color']].value_counts()

# Restaurant chain rating text count
text_Restaurant_Chains1=restaurant_Name.groupby(df[df['City']=='New Delhi']['City'])[['Restaurant Name','Rating text']].value_counts()

Cafe_Coffee_Day=[rating_Restaurant_Chains1,color_Restaurant_Chains1,text_Restaurant_Chains1]
print('\t-: Rating and popularity of Cafe Coffee Day restaurant chains in new delhi city :-')
for i in Cafe_Coffee_Day:
    print("\n",i)
print('\n\n')

#In Country:

restaurant_Name=df[df['Restaurant Name']=='Cafe Coffee Day']

# Restaurant chains rating count
rating_Restaurant_Chains1=restaurant_Name.groupby(df[df['Country Code']==1]['Country Code'])[['Restaurant Name','Aggregate rating']].value_counts()

# Restaurant chain rating color count
color_Restaurant_Chains1=restaurant_Name.groupby(df[df['Country Code']==1]['Country Code'])[['Restaurant Name','Rating color']].value_counts()

# Restaurant chain rating text count
text_Restaurant_Chains1=restaurant_Name.groupby(df[df['Country Code']==1]['Country Code'])[['Restaurant Name','Rating text']].value_counts()

Cafe_Coffee_Day=[rating_Restaurant_Chains1,color_Restaurant_Chains1,text_Restaurant_Chains1]
print('\t-: Rating and popularity of Cafe Coffee Day restaurant chains in country')
for i in Cafe_Coffee_Day:
    print("\n",i)

```
