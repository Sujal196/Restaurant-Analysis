Task 1: Top Cuisines

(A) Determine the top three most common cuisines in the dataset

```python
#Task 1
Top_cuisines=df['Cuisines'].value_counts().sort_values(ascending=False).head(3)
```

```python
Top_cuisines
```

 Determining the top three most common cuisines in the Dataset

```python
plt.figure(figsize=(5,5))
plt.plot(Top_cuisines,linewidth=0.7,color='blue',marker='o')
plt.title('Top Three Cuisines')
plt.xlabel("Cuisines Name")
plt.ylabel("Cuisines Count")
plt.ylim(200,1000)
plt.tight_layout()
plt.show()
```

Top Three Most Common Cuisines By Line Chart :

1. North Indian cuisines have high count more than 900.

2. North Indian, Chinese cuisines have count more than 500.

3. Fast Food have count more than 350.



(B) Calculate the percentage of restaurant that serves each of top cuisines

```python
Restaurant_count=len(df['Restaurant Name'])
Cuisines_count1=sum([i for i in df['Cuisines']=='North Indian'])
Cuisines_count2=sum([i for i in df['Cuisines']=='North Indian, Chinese'])
Cuisines_count3=sum([i for i in df['Cuisines']=='Fast Food'])
```

```python
#Percentage of Restaurant that serves North Indian food

Restaurant_pct1=(Cuisines_count1/Restaurant_count)*100
Restaurant_pct1
```

9.8% Restaurant that serves north indian food from top cuisines

```python
#Percentage of Restaurant that serves North Indian, Chinese

Restaurant_pct2=(Cuisines_count2/Restaurant_count)*100
Restaurant_pct2
```

5.3% Restaurant that serves north Indian, chinese food from top cuisines

```python
#Percentage of Restaurant that serves Fast Food
Restaurant_pct3=(Cuisines_count3/Restaurant_count)*100
Restaurant_pct3
```

3.7% Restaurant that serves fast food from top cuisines

```python
#Pie Chart that show Restaurant Percentage that serves top Cuisines
plt.figure(figsize=(5,7))
Restauran_percentage=[Restaurant_pct1,Restaurant_pct2,Restaurant_pct3]
Top_cuisines=['North Indian','North Indian, Chinese','Fast Food']
plt.pie(Restauran_percentage,explode=[0.01,0.01,0.01],labels=["North Indian","North Indian Chinese","Fast Foood"],autopct='%f1%%')
plt.title("Percentage of Restaurant Serves Top Cuisines")
plt.legend(loc='upper right',fontsize=6.9)
plt.tight_layout()
plt.show()
```

From Pie Chart :

1. 51.97 % of Restaurant that serves North Indian Food.
2. 28.37 % of Restaurant that serves North Indian, Chinese.
3. 19.65 % of Restaurant that serves Fast Food.

Most Restaurant Serves North Indian Food From Top Cuisines

Task 4: Online Delivery

(A) Determine the percentage of restaurants that offer online delivery.

```python
#Finding restaurant count that offer online delivery
Restaurant_online=df.groupby(df[df['Has Online delivery']=='Yes']['Has Online delivery'])['Restaurant Name'].count()
Restaurant_online
```

```python
#Counting total restaurant that offer online Delivery:
Total_Restaurant=len(df['Has Online delivery'])
Total_Restaurant
```

```python
#Calculating Percentage of restaurant that offers online delivery:
Restaurant_online_percentage=(Restaurant_online/Total_Restaurant)*100
Restaurant_online_percentage
```

25.66 %   of restaurants in dataset that offers online delivery.

(B)  Compare the average ratings of restaurants with and without online delivery.

```python
#Average Rating of restaurant that offers online delivery:
Restaurant_online_yes=df.groupby(df[df['Has Online delivery']=='Yes']['Has Online delivery'])['Aggregate rating'].mean()
Restaurant_online_yes
```

Average rating of restaurants that offers online delivery is 3.24

```python
#Average Rating of restaurant that offers no online delivery:
Restaurant_online_no=df.groupby(df[df['Has Online delivery']=='No']['Has Online delivery'])['Aggregate rating'].mean()
Restaurant_online_no
```

Average rating of restaurants that offers no online delivery is 2.46

```python
#Comparision of average rating of restaurant that offers online delivery or not:
plt.figure(figsize=(5,5))
sns.barplot(x=df['Has Online delivery'],y=df['Aggregate rating'],data=df,estimator='mean',color='purple',edgecolor='black',label="Average Rating")
plt.title("Comparision of average rating of restaurants")
plt.xlabel("Online Delivery")
plt.ylabel("Average Rating")
plt.legend(loc='upper right')
plt.ylim([1.0,4.0])
plt.tight_layout()
plt.show()
```

Average rating of restaurants that offers online delivery is more than average rating of restaurants that offers no online delivery.
